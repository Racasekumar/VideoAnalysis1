```python
import matplotlib.pyplot as plt
from detecto.utils import read_image
```

```python
base_path = "/content/drive/MyDrive/UpsolveAI/allimages"

from PIL import Image
# sample_image= "0.8_y1_Color.png"
im = Image.open("0.8_y1_Color.png")
sample_image = im.convert('RGB')
sample_image.save('colors.jpg')





image = read_image(base_path + sample_image)
plt.imshow(image)
plt.show()
```

```python
from detecto.core import Dataset
dataset = Dataset(base_path)
```

```python
from detecto.visualize import show_labeled_image

image, targets = dataset[30]
show_labeled_image(image, targets["boxes"], targets["labels"])
```

```python
from detecto.core import DataLoader, Model

labels = ["table","bottle",
          "box","handwash",
          "toiletpaper","screw",
          "washer","noodlescup",
          "chips","shoes",
          "sanitary_products","snacks",
          "paper","human",
          "www"]
model = Model(labels)
```

```python
model.fit(dataset)
```

```python
image = read_image(base_path+"0.8_o3_Color.png")
labels, boxes, scores = model.predict(image)
```

```python
print["labels", labels]
print["boxes", boxes]
print["scores", scores]
```

```python
from detecto.visualize import show_labeled_image
show_labeled_image(image, boxes[0], labels[0])
```
